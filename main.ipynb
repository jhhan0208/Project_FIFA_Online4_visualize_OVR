### 원본 파일 불러오기

```python
import pandas as pd

data = pd.read_csv('data.csv')
#data.head(20)
```

### 사이트에서 팀 정보 획득

```python
select_team = []
team_data = pd.read_csv('find_team.csv') 
for i in range(0, len(team_data), 10):
    select_team.append(team_data['name'][i])
    
select_team = pd.DataFrame({'name': select_team})
```

### 1차 업데이트

```python
player_name = []
first_ovr = []
first_dif = []
second_ovr = []

for i in range(0, 3970, 10):
    player_name.append(data['players'][i])
    if(len(data['players'][i+4]) > 3):
        x = data['players'][i+4]
        a, _, B = x.partition('(')
        b = B[:-1]
        first_ovr.append(int(b))
        first_dif.append(int(a)-int(b))
        second_ovr.append(int(a))
    else:
        first_ovr.append(int(data['players'][i+4]))
        first_dif.append(0)
        second_ovr.append(int(data['players'][i+4]))
    
first_list = pd.DataFrame({'name': player_name})
```

### 팀 정보 추가

```python
teams = []
for i in range(0, 397):
    if(first_list['name'][i] in list(pd.read_csv('teams\맨체스터 시티.txt')['name'])): teams.append('맨체스터 시티')
    elif(first_list['name'][i] in list(pd.read_csv('teams\첼시.txt')['name'])): teams.append('첼시')    
    elif(first_list['name'][i] in list(pd.read_csv('teams\아틀레티코 마드리드.txt')['name'])): teams.append('아틀레티코 마드리드')  
    elif(first_list['name'][i] in list(pd.read_csv('teams\레알 마드리드.txt')['name'])): teams.append('레알 마드리드')  
    elif(first_list['name'][i] in list(pd.read_csv('teams\파리 생제르맹.txt')['name'])): teams.append('파리 생제르맹') 
    elif(first_list['name'][i] in list(pd.read_csv('teams\바이에른 뮌헨.txt')['name'])): teams.append('바이에른 뮌헨') 
    elif(first_list['name'][i] in list(pd.read_csv('teams\FC 포르투.txt')['name'])): teams.append('FC 포르투') 
    elif(first_list['name'][i] in list(pd.read_csv('teams\보루시아 도르트문트.txt')['name'])): teams.append('보루시아 도르트문트') 
    elif(first_list['name'][i] in list(pd.read_csv('teams\FC 바르셀로나.txt')['name'])): teams.append('FC 바르셀로나') 
    elif(first_list['name'][i] in list(pd.read_csv('teams\리버풀.txt')['name'])): teams.append('리버풀')
    elif(first_list['name'][i] in list(pd.read_csv('teams\라치오.txt')['name'])): teams.append('라치오')
    elif(first_list['name'][i] in list(pd.read_csv('teams\보루시아 묀헨글라트바흐.txt')['name'])): teams.append('보루시아 묀헨글라트바흐')
    elif(first_list['name'][i] in list(pd.read_csv('teams\세비야 FC.txt')['name'])): teams.append('세비야 FC')
    elif(first_list['name'][i] in list(pd.read_csv('teams\아탈란타.txt')['name'])): teams.append('아탈란타')
    elif(first_list['name'][i] in list(pd.read_csv('teams\RB 라이프치히.txt')['name'])): teams.append('RB 라이프치히')
    elif(first_list['name'][i] in list(pd.read_csv('teams\유벤투스.txt')['name'])): teams.append('유벤투스')
    else: teams.append('nan')

first_list = pd.DataFrame({'name': player_name, 'team': teams, 'first_ovr': first_ovr, 
                           'first_dif': first_dif, 'second_ovr': second_ovr})
```

### 1차 업데이트 + 2차 업데이트 + 3차 업데이트 + 4차 업데이트

```python
first_list_combine = first_list.sort_values(by = 'name', ascending = True)
second_list_combine = second_list.sort_values(by = 'name', ascending = True)
third_list_combine = third_list.sort_values(by = 'name', ascending = True)
fourth_list_combine = fourth_list.sort_values(by = 'name', ascending = True)

first_list_combine = first_list_combine.reset_index(drop=True)
second_list_combine = second_list_combine.reset_index(drop=True)
third_list_combine = third_list_combine.reset_index(drop=True)
fourth_list_combine = fourth_list_combine.reset_index(drop=True)

second_dif = []
third_ovr = []

third_dif = []
fourth_ovr = []

fourth_dif = []
fifth_ovr = []

for i in range(0, 397):
    second_dif.append(int(second_list_combine['third_ovr'][i]) - int(first_list_combine['second_ovr'][i]))
    third_ovr.append(int(second_list_combine['third_ovr'][i]))
    third_dif.append(int(third_list_combine['fourth_ovr'][i]) - int(second_list_combine['third_ovr'][i]))
    fourth_ovr.append(int(third_list_combine['fourth_ovr'][i]))
    fourth_dif.append(int(fourth_list_combine['fifth_ovr'][i]) - int(third_list_combine['fourth_ovr'][i]))
    fifth_ovr.append(int(fourth_list_combine['fifth_ovr'][i]))
    
first_list_combine['second_dif'] = second_dif
first_list_combine['third_ovr'] = third_ovr

first_list_combine['third_dif'] = third_dif
first_list_combine['fourth_ovr'] = fourth_ovr

first_list_combine['fourth_dif'] = fourth_dif
first_list_combine['fifth_ovr'] = fifth_ovr

final_list = first_list_combine.sort_values(by = 'fifth_ovr', ascending = False)

ind = []
for i in range(0, 397):
    ind.append(i)

final_list.index = ind

final_list.head()
```

첼시 선수들 전체 오버롤 변화 시각화

```python
x = first_list_combine.sort_values(by = 'fifth_ovr', ascending = False)

y = x.loc[(x['team'] == '첼시')]

z = pd.DataFrame({'name': y['name'], 'first_ovr': y['first_ovr'],'second_ovr': y['second_ovr'], 
                  'third_ovr': y['third_ovr'], 'fourth_ovr': y['fourth_ovr'], 'fifth_ovr': y['fifth_ovr']})

l1 = list(z['first_ovr'])
l2 = list(z['second_ovr'])
l3 = list(z['third_ovr'])
l4 = list(z['fourth_ovr'])
l5 = list(z['fifth_ovr'])

plt.figure(figsize=(20, 10))

for i in range(0, len(l1)):
    plt.plot(['first_ovr', 'second_ovr', 'third_ovr', 'fourth_ovr', 'fifth_ovr'], [int(l1[i]), int(l2[i]), int(l3[i]), int(l4[i]), int(l5[i])], 'o:')
    plt.legend(z['name'])
    

plt.show()
```

```python
x = first_list_combine.sort_values(by = 'fifth_ovr', ascending = False)

#y = x.loc[(x['team'] == '첼시') & (x['fourth_ovr'] > 95)]

y = x.loc[(x['name'] == '메이슨 마운트') | (x['name'] == '크리스천 풀리식') | (x['name'] == '카이 하베르츠')]

#y = x.loc[(x['name'] == '마르코 아센시오')]

z = pd.DataFrame({'name': y['name'], 'first_ovr': y['first_ovr'],'second_ovr': y['second_ovr'], 
                  'third_ovr': y['third_ovr'], 'fourth_ovr': y['fourth_ovr'], 'fifth_ovr': y['fifth_ovr']})

l1 = list(z['first_ovr'])
l2 = list(z['second_ovr'])
l3 = list(z['third_ovr'])
l4 = list(z['fourth_ovr'])
l5 = list(z['fifth_ovr'])

plt.figure(figsize=(5, 5))

for i in range(0, len(l1)):
    plt.plot(['first_ovr', 'second_ovr', 'third_ovr', 'fourth_ovr', 'fifth_ovr'], [int(l1[i]), int(l2[i]), int(l3[i]), int(l4[i]), int(l5[i])], 'o:')
    plt.legend(z['name'])
    

plt.show()
```
